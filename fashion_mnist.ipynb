```python
X_train=X_train/255.0
X_test=X_test/255.0
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
  tf.keras.layers.Dense(10,activation="softmax"),
    ])
```

```python
model.compile(loss="SparseCategoricalCrossentropy",optimizer="adam",metrics="accuracy")
```

```python
norm_hist=model.fit(X_train,y_train,epochs=15,validation_data=(X_test,y_test))
```

```python
import pandas as pd
pd.DataFrame(hist.history).plot(title="Non-Normalized")
pd.DataFrame(norm_hist.history).plot(title="Normalized")
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
  tf.keras.layers.Dense(10,activation="softmax"),
    ])


model.compile(loss="SparseCategoricalCrossentropy",optimizer="adam",metrics="accuracy")

lr_sd=tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-3 * 10**(epoch/20))

lr_hist=model.fit(X_train,y_train,epochs=40,validation_data=(X_test,y_test))
```

```python
lrs=1e-3 *(10**(tf.range(40)/20))
plt.semilogx(lrs,lr_hist.history["loss"])
plt.title("Learning rate")
plt.xlabel("Learning rate")
plt.ylabel("Loss")
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
  tf.keras.layers.Dense(10,activation="softmax"),
    ])


model.compile(loss="SparseCategoricalCrossentropy",optimizer=tf.keras.optimizers.Adam(lr=0.1),metrics="accuracy")


updated_lr_hist=model.fit(X_train,y_train,epochs=40,validation_data=(X_test,y_test))
```
